# LOTZ

```python
Y = [3494, 13480, 52999, 223346]
Yfixed = [5419, 23518, 96294, 382704] #Averages of 100 experiments
Yfixedv2 = [6103, 28579, 124061, 555993] #Avgs of 100 experiments

Y2inf = [5405.59, 24340.6, 98512.2, 382678]
Y1inf = [8631.34, 41791.2, 175387, 723943]
fig = plt.figure(figsize=(10,10))
plt.plot(X[0:len(Y)], Y, marker="o", label="NSGA-II (bug)")
# plt.plot(X[0:len(Yfixed)], Yfixed, marker="o", color="black", label="NSGA-II")
# plt.plot(X[0:len(Yfixedv2)], Yfixedv2, marker="o", color="red", label="NSGA-II v2")
plt.plot(X[0:len(Y2inf)], Y2inf, marker="o", color="green", label="NSGA-II (2 infinites per objective)")
plt.plot(X[0:len(Y1inf)], Y1inf, marker="o", color="orange", label="NSGA-II (1 infinite per objective)")
plt.yscale('log')
plt.xscale('log')
plt.xticks(X[0:len(Y)], ["25", "50", "100", "200"])
plt.grid()
plt.ylim(10**2, 10**7)
plt.legend()
```
