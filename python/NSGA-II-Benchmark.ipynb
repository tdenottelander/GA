```python
import matplotlib.pyplot as plt
```

```python
X = [25, 50, 100, 200, 400]
```

# LOTZ

```python
Y = [3494, 13480, 52999, 223346]
fig = plt.figure(figsize=(10,10))
plt.plot(X[0:len(Y)], Y, marker="o")
plt.yscale('log')
plt.xscale('log')
plt.xticks(X[0:len(Y)], ["25", "50", "100", "200"])
plt.grid()
plt.ylim(10**2, 10**7)
```
